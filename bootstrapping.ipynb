```python
import numpy as np
import scipy.stats as sp
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n = 10_000
original_data = np.random.lognormal(1, 2, size=n)
```

```python
boostrapped_means = []
for i in range(5_000):
    bootstrap_mean = np.random.choice(original_data, size=n).mean()
    
    boostrapped_means.append(bootstrap_mean)
```

```python
sns.kdeplot(original_data)
plt.show()

sns.kdeplot(boostrapped_means)
plt.show()
```
