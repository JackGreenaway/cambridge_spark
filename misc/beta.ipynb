```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
df = yf.download(["MSFT", "NVDA", "META", "GS", "TSLA", "AMZN", "AAPL", "^SPX"], start="2015-01-01")["Adj Close"]
```

```python
pct = df.pct_change().dropna()
```

```python
pct
```

```python
m_rets = (df["^SPX"][-1] - df["^SPX"][0]) / df["^SPX"][0]

beta = pd.DataFrame(columns=["beta", "rets-m_rets"])
for asset in df.columns[:-1]:    
    
    target = [asset, "^SPX"]
    rets = (df[asset][-1] - df[asset][0]) / df[asset][0]
    
    cov = pct[target].cov().values[0][1] 

    b = cov / np.var(pct["^SPX"].values)
    
    perf = (rets - m_rets) * 100
    beta.loc[asset] = [b.round(2), perf]
    # print(f"beta of {asset}: {b:.2f}")

beta.sort_values(ascending=False, by="beta")
```
