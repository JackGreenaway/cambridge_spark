```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.uniform(0, 20, 50)
y = 2 * x + 5 + np.random.normal(0, 3, 50)
```

```python
fig, ax = plt.subplots()

colour = "#5795cf"
ax.set_facecolor(colour)
fig.set_facecolor(colour)

ax.tick_params(axis='x', colors='#fcba03')
ax.tick_params(axis='y', colors='#fcba03')

ax.spines[:].set_color("#eb348c")

ax.scatter(x, y, color="w")
```
