```python
import pandas as pd
import numpy as np

# Creating a larger sample DataFrame
np.random.seed(42)

data = {'feature1': np.random.choice(['A', 'B', 'C', 'D', 'E'], size=200),
        'feature2': np.random.choice(['X', 'Y', 'Z'], size=200),
        'variable': np.random.choice(['Cat', 'Dog', 'Bird', 'Fish', 'Rabbit', 'Snake'], size=200)}

df = pd.DataFrame(data)

# Group by 'feature1' and 'feature2', then calculate the value counts for 'variable'
grouped_df = df.groupby(['feature1', 'feature2'])['variable'].value_counts().unstack(fill_value=0)

# Calculate the percentage value counts
percentage_df = grouped_df.div(grouped_df.sum(axis=1), axis=0) * 100

# Print the result
percentage_df
```
